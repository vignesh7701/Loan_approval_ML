```python
df1['AMT_ANNUITY'].fillna(df1['AMT_ANNUITY'].mean(),inplace=True)
df1['AMT_GOODS_PRICE']=df1['AMT_GOODS_PRICE'].fillna(0)
df1['NAME_TYPE_SUITE']=df1['NAME_TYPE_SUITE'].fillna('Unaccompanied')
df1['CNT_PAYMENT']=df1['CNT_PAYMENT'].fillna(0)
df1['DAYS_FIRST_DUE']=df1['DAYS_FIRST_DUE'].fillna(0)
df1['DAYS_LAST_DUE_1ST_VERSION']=df1['DAYS_LAST_DUE_1ST_VERSION'].fillna(0)
df1['DAYS_LAST_DUE']=df1['DAYS_LAST_DUE'].fillna(0)
df1['DAYS_TERMINATION']=df1['DAYS_TERMINATION'].fillna(0)
df1['NFLAG_INSURED_ON_APPROVAL']=df1['NFLAG_INSURED_ON_APPROVAL'].fillna(df1['NFLAG_INSURED_ON_APPROVAL'].mean())
df1['PRODUCT_COMBINATION']=df1['PRODUCT_COMBINATION'].fillna('Cash')
df1['DAYS_FIRST_DRAWING'].fillna(df1['DAYS_FIRST_DRAWING'].mean(),inplace=True)
df1['RATE_DOWN_PAYMENT'].fillna(df1['RATE_DOWN_PAYMENT'].mean(),inplace=True)
df1['AMT_DOWN_PAYMENT']=df1['AMT_DOWN_PAYMENT'].fillna(df1['AMT_DOWN_PAYMENT'].mean())
```

```python
cols = ['SK_ID_PREV', 'SK_ID_CURR', 'NAME_CONTRACT_TYPE', 'AMT_ANNUITY',
                    'AMT_APPLICATION', 'AMT_CREDIT', 'AMT_DOWN_PAYMENT', 'AMT_GOODS_PRICE',
                    'WEEKDAY_APPR_PROCESS_START', 'HOUR_APPR_PROCESS_START',
                    'FLAG_LAST_APPL_PER_CONTRACT', 'NFLAG_LAST_APPL_IN_DAY',
                    'RATE_DOWN_PAYMENT',
                     'NAME_CASH_LOAN_PURPOSE',
                    'NAME_CONTRACT_STATUS', 'DAYS_DECISION', 'NAME_PAYMENT_TYPE',
                    'CODE_REJECT_REASON', 'NAME_TYPE_SUITE', 'NAME_CLIENT_TYPE',
                    'NAME_GOODS_CATEGORY', 'NAME_PORTFOLIO', 'NAME_PRODUCT_TYPE',
                    'CHANNEL_TYPE', 'SELLERPLACE_AREA', 'NAME_SELLER_INDUSTRY',
                    'CNT_PAYMENT', 'NAME_YIELD_GROUP', 'PRODUCT_COMBINATION',
                    'DAYS_FIRST_DRAWING', 'DAYS_FIRST_DUE', 'DAYS_LAST_DUE_1ST_VERSION',
                    'DAYS_LAST_DUE', 'DAYS_TERMINATION', 'NFLAG_INSURED_ON_APPROVAL']

imbalance_ratios = {}

for column in cols:
    class_counts = df1[column].value_counts()
    if len(class_counts) > 1:
        imbalance_ratio = class_counts.iloc[0] / class_counts.iloc[1]
    else:
        imbalance_ratio = 'Not Applicable'
    imbalance_ratios[column] = imbalance_ratio

# Print the imbalance ratios
for column, ratio in imbalance_ratios.items():
    print(f"Imbalance Ratio for '{column}': {ratio}")
```
